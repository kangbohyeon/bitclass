## p81 연습

```python
sel = int(input("입력 진수 결정(16/10/8/2) : "))
num = input("값 입력 : ")
```

```python
sel, num
```

```python
if sel == 16:
    num10 = int(num, 16) 
if sel == 10 :
    num10 = int(num, 10)
if sel == 8 :
    num10 = int(num, 8)
if sel == 2 :
    num10 = int(num, 2)
```

```python
print("16진수 ==> ", hex(num10))
print("10진수 ==> ", num10)
print(" 8진수 ==> ", oct(num10))
print(" 2진수 ==> ", bin(num10))
```

#### 문16

> 제어문 학습후 풀기

```python
num=input(" 글자 입력 : ")
if ('0' <= num and num <= '1') :
    print(" 2진수 또는 8진수 또는 10진수 또는 16진수 입니다")
elif ('2' <= num and num <= '7') :
    print(" 8진수 또는 10진수 또는 16진수 입니다")
elif ('8' <= num and num <= '9') :
    print(" 10진수 또는 16진수 입니다")
elif ('a' <= num and num <= 'f') or ('A' <= num and num <= 'F') :
    print(" 16진수 입니다")
else :
    print(" 숫자가 아닙니다.")
```
