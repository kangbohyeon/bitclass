생성자 추가

```python
class Car :
    # 속성: 필드
    color = ""
    speed = 0
    
    # 생성자
    def __init__(self):
        self.color = '빨강'
        self.speed = 0
    
    # 메서드
    def upspeed(self, value):
        self.speed += value
        
    def downspeed(self, value):
        self.speed -= value
    
    def print(self):
        print("Car class")
```

```python
car1 = Car()
car2 = Car()

print(car1.color, car2.color)
```

매개변수가 있는 생성자!!!

```python
class Car :
    # 속성: 필드
    color = ""
    speed = 0
    
    # 생성자
    def __init__(self, v1, v2):
        self.color = v1
        self.speed = v2
    
    # 메서드
    def upspeed(self, value):
        self.speed += value
        
    def downspeed(self, value):
        self.speed -= value
    
    def print(self):
        print(self.color, self.speed)
```

```python
car1 = Car("노랑", 0)
car2 = Car("오렌지", 0)
```

```python
car1.print(), car2.print()
```

인스턴스 변수/클래스 변수 (파이썬)

```python
class Car :
    # 속성: 필드
    color = ""
    speed = 0
    count = 0   # 클래스 변수(영역)
    
    # 생성자
    def __init__(self):
        self.speed = 0
        Car.count += 1
    
    # 메서드
    def upspeed(self, value):
        self.speed += value
        
    def downspeed(self, value):
        self.speed -= value
    
    def print(self):
        print(self.color, self.speed)
```

```python
car1 = Car()
car1.speed = 30
```

```python
car1.speed
```

```python
car1.count
```

```python
Car.count
```

```python
car2 = Car()
car2.speed = 60
```

```python
car2.speed
```

```python
car2.count
```

```python
car2.count = 1000
```

상속

```python
class Car:
    speed = 0
    
    def upspeed(self, v):
        self.speed += v
        print('속도', self.speed)
```

```python
class Sedan(Car):
    def upspeed(self, v):
        self.speed += v
        
        if self.speed > 150:
            self.speed = 150
```

```python
class Truck(Car):
    pass
```

```python
truck1 = Truck()
sedan1 = Sedan()
```

```python
truck1.upspeed(100)
```
