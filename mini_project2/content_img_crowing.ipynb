```python
# riss.kr 에서 특정 키워드로 논문 / 학술 자료 검색하기

#Step 1. 필요한 모듈을 로딩합니다
import time          
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
from selenium.webdriver.common.keys import Keys


path = '/Users/bohyenkang/Library/Mobile Documents/com~apple~CloudDocs/coding/bitclass/mini_project1/second_data/second_total_data.csv'
temp = pd.read_csv(path)
df = temp.drop(temp.columns[[0]],axis=1)

# chrome을 전체화면으로 넓히는 옵션입니다.
options = Options()
options.add_argument('--start-fullscreen')

#Step 3. 크롬 드라이버 설정 및 웹 페이지 열기
chrome_path = "/Users/bohyenkang/Downloads/chromedriver"
#driver=webdriver.Chrome(ChromeDriverManager().install())
driver = webdriver.Chrome(executable_path=chrome_path, chrome_options=options)

url = 'https://www.kmdb.or.kr/db/kor/detail/movie/F/07207'
driver.get(url)
time.sleep(2)


#영화 제목 추출
a = [] #검색결과 없는 인덱스
b = [] #url
c = [] #줄거리
d = [] # 영화제목
#Step 2. 사용자에게 검색 관련 정보들을 입력 받습니다.
try :
    for i in range(10):
    #     print("=" *100)
    #     print(" 이 크롤러는 RISS 사이트의 논문 및 학술자료 수집용 웹크롤러입니다.")
    #     print("=" *100)
        #query_txt = input('1.수집할 자료의 키워드는 무엇입니까?(여러개일 경우 , 로 구분하여 입력): ')

        query_txt = df.loc[i][0]
        query_year = df.loc[i][2]
        #print("\n")
        #상세검색 click
        driver.find_element_by_xpath('//*[@id="header"]/header/div[1]/div/div/a').click()

    #     #영화제목 click

        element = driver.find_element_by_xpath('//*[@id="labelName"]')
        element.click()
        element.send_keys(query_txt)

        #제작년도 시작 click
        element = driver.find_element_by_xpath('//*[@id="labelYear"]')
        element.click()
        element.send_keys((str)(query_year))
        #제작년도 종료 click
        element = driver.find_element_by_xpath('//*[@id="fieldset"]/div/dl/dd[9]/div[3]/span[1]/input')
        element.click()
        element.send_keys((str)(query_year))

        #검색 click
        element = driver.find_element_by_xpath('//*[@id="fieldset"]/div/div/span[2]/a')
        element.click()
        time.sleep(2)


        #검색결과가 없을시 예외처리
        try :
            link = driver.find_element_by_class_name('movie-tt')
            link.click()
            time.sleep(2)
        except : 
            a.append(i)
            continue

        driver.find_element_by_class_name('iOpen1').click() #펼쳐보기 클릭

        link = driver.find_element_by_xpath('//*[@id="fieldset"]/div[5]/div[2]') #줄거리 경로


        bsObject = BeautifulSoup(driver.page_source, 'html.parser') #이미지 url 
        time.sleep(1)
        pr_img = bsObject.find_all(class_='mImg1')
        img_url = str(pr_img[0]).split('>')[1].split(' ')[2].split("'")[1]
        b.append(img_url)
        c.append(link.text)
        d.append(query_txt)
        print(i)
    #     print()
    #     print(query_txt)
    #     print(link.text)
    #     print()
    data = {
        'title' : d,
        'img_url': b,
        'content': c
    }

    df = pd.DataFrame(data)
    df.to_csv('temp.csv', index=True)

    data = {
        'non_index' : a
    }

    df = pd.DataFrame(data)
    df.to_csv('index_temp.csv', index=True)
except : 
    data = {
        'title' : d,
        'img_url': b,
        'content': c
    }

    df = pd.DataFrame(data)
    df.to_csv('temp.csv', index=True)

    data = {
        'non_index' : a
    }

    df = pd.DataFrame(data)
    df.to_csv('index_temp.csv', index=True)
```
